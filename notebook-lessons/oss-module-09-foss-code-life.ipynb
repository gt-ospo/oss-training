## 1: Natural Open Source Lifecycle üîÑ
### Open Source Project Lifecycle: Project ‚Üí Ecosystem ‚Üí Archive

#### Project Phase
- Initial development, usually by one or a few individuals.
- Focused scope, solving a specific problem.
- Basic integrations appear (CI/CD, documentation generation).
- Examples: first commit, README, LICENSE.

#### Ecosystem Phase
- Community forms, contributions increase.
- Features grow, broader adoption occurs.
- Tools like GitHub Issues, Discussions, and Projects are used for community management.

#### Archive Phase
- Maintainers shift priorities, activity drops.
- Project is deprecated, forked, or archived.
- A good exit strategy includes archive notices and project transfer guidance.

---

## 2. Sustainability Practices for Each Lifecycle Phase üå±

### Project Phase
- **Actionable Steps**:
  - Define a clear project scope and goals.
  - Create essential files: `README.md`, `LICENSE`, and `CODE_OF_CONDUCT.md`.
  - Use GitHub Issues to track bugs and features.
  - Set up basic automation with GitHub Actions for CI/CD.

### Ecosystem Phase
- **Actionable Steps**:
  - Foster community engagement through GitHub Discussions.
  - Expand automation workflows and documentation generation.
  - Provide detailed documentation using generation tools (MkDocs, Sphinx) and hosting platforms (ReadTheDocs, GitHub Pages).

### Archive Phase
- **Actionable Steps**:
  - Announce deprecation with clear archive notices.
  - Provide guidance for forks and project transfers.
  - Maintain transparency with contributors and users.

## 4. GitHub Sustainability Analyzer Tool üîé

```python
# Let's Check Your Open Source Project Sustainability!
# This script analyzes a GitHub repository for sustainability indicators.
# <--- Instructions: --->
# First, run this cell to define the functions needed for analysis.
# Then run the cell below to test it with a specific repository URL!

import requests
from urllib.parse import urlparse

GITHUB_API = "https://api.github.com"

def extract_repo_info(url):
    """Extract user and repo name from GitHub URL"""
    path = urlparse(url).path.strip("/").split("/")
    if len(path) >= 2:
        return path[0], path[1]
    raise ValueError("Invalid GitHub repository URL")

def fetch_github_repo_metadata(user, repo):
    """Fetch repository metadata from GitHub API"""
    repo_api = f"{GITHUB_API}/repos/{user}/{repo}"
    response = requests.get(repo_api)
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"Failed to fetch metadata: {response.status_code}")

def check_file_exists(user, repo, file_path):
    """Check if a specific file exists in the repository"""
    url = f"{GITHUB_API}/repos/{user}/{repo}/contents/{file_path}"
    r = requests.get(url)
    return r.status_code == 200

def analyze_repo(url):
    """Analyze a GitHub repository for sustainability indicators"""
    user, repo = extract_repo_info(url)
    metadata = fetch_github_repo_metadata(user, repo)
    
    indicators = {
        "Has License": check_file_exists(user, repo, "LICENSE"),
        "Has Readme": check_file_exists(user, repo, "README.md"),
        "Has Contributing Guide": check_file_exists(user, repo, "CONTRIBUTING.md"),
        "Has Code of Conduct": check_file_exists(user, repo, "CODE_OF_CONDUCT.md"),
        "Has GitHub Actions": check_file_exists(user, repo, ".github/workflows"),
        "Open Issues": metadata.get("open_issues_count", 0),
        "Forks": metadata.get("forks_count", 0),
        "Stars": metadata.get("stargazers_count", 0),
        "Archived": metadata.get("archived", False),
        "Last Updated": metadata.get("updated_at"),
        "Community Engagement": metadata.get("subscribers_count", 0)
    }

    return indicators

def test_analyze_repo(repo_url):
    """Test the analyze_repo function with a provided GitHub URL"""
    indicators = analyze_repo(repo_url)
    print("--- Sustainability Indicators ---")
    
    # Define which metrics are boolean vs numeric
    boolean_metrics = {
        "Has License", "Has Readme", "Has Contributing Guide", 
        "Has Code of Conduct", "Has GitHub Actions", "Archived"
    }
    
    for key, value in indicators.items():
        if key in boolean_metrics:
            # For boolean metrics, use standard boolean logic
            emoji = "‚úÖ" if value else "‚ùå"
        elif key == "Archived":
            # Archived should be ‚ùå if True (archived is bad for active projects)
            emoji = "‚ùå" if value else "‚úÖ"
        elif isinstance(value, (int, float)):
            # For numeric metrics, show actual value without emoji judgment
            emoji = "üìä"
        else:
            # For other types (like dates), just show the value
            emoji = "üìÖ"
        
        print(f"{emoji} {key}: {value}")
```

```python
repo_url = "https://github.com/gt-ospo/oss-training"
test_analyze_repo(repo_url)
```

## 5. Research Foundations for OSS Sustainability üî¨

### Understanding Dependency Abandonment and Supply Chain Trust

Recent software engineering research has shed light on critical challenges in open source ecosystem sustainability. As open source projects grow, they become increasingly interdependent, creating complex supply chains where the abandonment of a single dependency can have cascading effects.

#### Key Research Areas

##### 1. When Does Dependency Abandonment Matter?
**Miller, et al. "Designing Abandabot: When Does Open Source Dependency Abandonment Matter?"** *ICSE 2026*

This research investigates:
- **Triggers of Abandonment**: What causes maintainers to stop actively developing a project?
- **Risk Assessment**: Which abandoned dependencies pose the greatest risk to downstream projects?
- **Early Warning Systems**: How can we identify projects at risk of abandonment before it happens?
- **Mitigation Strategies**: What approaches help projects survive when their dependencies are abandoned?

##### 2. Trust in Open Source Supply Chains
**Boughton, et al. "Decomposing and Measuring Trust in Open-Source Software Supply Chains"** *ICSE-NIER 2024*

This work provides frameworks for understanding and measuring trust:
- **Trust Factors**: Transparency, security practices, community engagement, and maintainer reliability
- **Supply Chain Visibility**: How to identify and evaluate the trustworthiness of dependencies
- **Risk Evaluation**: Assessing the stability and health of projects you depend on
- **Governance Impact**: How different governance models affect trust and reliability

##### 3. Ecosystem Response to Abandonment  
**Miller, et al. "Understanding the Response to Open-Source Dependency Abandonment in the npm Ecosystem"** *ICSE 2025*

This study analyzes real-world responses in the npm ecosystem:
- **Community Adaptations**: How projects adapt when dependencies are abandoned (forking, vendoring, migration)
- **Response Timeline**: How quickly does the ecosystem respond to abandonment?
- **Success Factors**: What determines whether a project successfully navigates dependency abandonment?
- **Ecosystem Health**: Metrics for measuring ecosystem resilience to abandonment

#### Practical Implications for Your Projects

When evaluating OSS dependencies and building sustainable projects:

1. **Monitor Dependency Health**: Use tools to track the activity level and health of upstream dependencies
2. **Plan for Alternatives**: Identify backup solutions or forks for critical dependencies
3. **Engage with Maintainers**: Build relationships and communicate about dependency stability
4. **Document Assumptions**: Make explicit what versions and behaviors your project depends on
5. **Support Maintainers**: Consider supporting the maintenance of critical dependencies through funding or contributions
6. **Plan for Succession**: Ensure your own project has plans for graceful transition if abandonment occurs

### Choosing a License

Selecting the right license is crucial for the success of your open source project!

- **For Corporate Use**: Apache 2.0 license is often preferred due to its permissive nature and explicit patent grant, fostering trust and legal clarity.
- **For Copyleft Projects**: GNU GPL enforces strong copyleft (all future modifications must be open source), while LGPL offers more flexibility (allows linking with proprietary software).
- **For Academic Use**: Permissive licenses (MIT, BSD) are often the best choice.
- **For Non-code Projects**: Creative Commons licenses are recommended.

You must choose a license that aligns with your project's goals and community values! If you're unsure, consider using the [ChooseALicense.com](https://choosealicense.com/) tool to help you select the right license for your project - and reach out to [GT's OSPO for guidance](https://ospo.cc.gatech.edu/)!
